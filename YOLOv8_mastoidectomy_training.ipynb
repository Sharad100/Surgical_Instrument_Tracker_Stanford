```python
from PIL import Image
from ultralytics import YOLO
import numpy as np
import pandas as pd
import torch

import os
import re

directory = '/content/drive/Shareddrives/Share-with-Sharad/Zavala_george'  # Replace with the actual directory path

# Get a list of files in the directory
file_list = os.listdir(directory)

# Filter for files that match the naming pattern (e.g., 'vlc_extractedXXXXX.jpeg')
original_files = [filename for filename in file_list if re.match(r'^vlc_extracted\d+\.jpeg$', filename)]

# Define a custom sorting key to extract and sort by the numeric part
def extract_numeric_part(filename):
    return int(filename.split('extracted')[1].split('.')[0])

# Sort the files based on the numeric part of the filename
sorted_files = sorted(original_files, key=extract_numeric_part)


model = YOLO('/content/drive/MyDrive/YOLOV8/training_results/train/weights/best.pt')
df = pd.DataFrame(columns=['frame_name', 'class_label', 'x', 'y', 'w', 'h'])



#for image_path in os.listdir('/content/drive/MyDrive/YOLOV8/data/images/test'):
for filename in sorted_files:
    path = '/content/drive/Shareddrives/Share-with-Sharad/Zavala_george/' + filename
    results = model(path)
    for r in results:
      new_boxes = []
      for box in r.boxes:
        if (box.cls.item() == 0.0 and box.conf.item() > 0.6) or (box.cls.item() == 1.0 and box.conf.item() > 0.25):
          new_boxes.append(box)
      r.boxes = new_boxes
      for box in r.boxes:
        class_label = model.names[int(box.cls)]
        x = box.xywh[0][0].item()
        y = box.xywh[0][1].item()
        w = box.xywh[0][2].item()
        h = box.xywh[0][3].item()
        #print("x = ", x)
        #print("y = ", y)
        #print("w = ", w)
        #print("h = ", h)
        confidence_score = box.conf.item()
        #("confidence_score = ", confidence_score)
        df = df.append({'frame_name': filename,
                        'class_label': class_label,
                        'x': x,
                        'y': y,
                        'w': w,
                        'h': h,
                        'confidence_score': confidence_score}, ignore_index=True)
      im_array = r.plot()  # plot a BGR numpy array of predictions
      im = Image.fromarray(im_array[..., ::-1])  # RGB PIL image
      im.show()  # show image
      im.save('/content/drive/Shareddrives/Share-with-Sharad/Zavala_george_output_frames_drill_conf_0.6_suction_conf_0.25/' + filename)  # save image

df.to_excel('/content/drive/MyDrive/YOLOV8/bounding_boxes_coords_Zavala_george_set.xlsx', index=False)








```
